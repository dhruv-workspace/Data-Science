**Valid Sudoku**

```python
import numpy as np
```

```python
s = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],
    
    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],
    
    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
])

row_wise = np.sum(s,axis=0)
column_wise = np.sum(s,axis=1)
if np.array_equal(row_wise,column_wise):
    print('Sudoku is valid')
else:
    print('Sudoku is not valid')
```

```python
import numpy as np

# Columns: [Age, Math Marks, Science Marks]
data = np.array([
    [18, 85, 78],   # Student 1
    [19, 92, 88],   # Student 2
    [17, 76, 95],   # Student 3
    [18, 65, 70],   # Student 4
    [20, 90, 85]    # Student 5
])
```

```python
# Get the shape of the matrix.
data.shape
```

```python
# Find the average age of students.
sum_age = data[0:,0].sum()
avg_age = sum_age / 5
print(avg_age)
# Direct mean function
print(np.mean(data[0:,0]))
```

```python
# Extract Math marks of all students.
math_marks = data[0:,1]
print(math_marks)
```

```python
# Find the highest Science mark.
science_marks = data[0:,2]
print(science_marks.max())
```

```python
# Get details of the student who scored more than 90 in Math.
data[data[:,1] > 90]
```

```python
# Increase math marks of all students with 5
data = data[:,1] + 5
data
```

```python
# Find how many students are younger than 19.
age = data[:,0]
bool_age = age < 19
print(len(age[bool_age]))
```

```python
# Calculate the average marks in each subject (column-wise mean).
maths_avg = np.mean(math_marks)
science_avg = np.mean(science_marks)
print(math_avg, science_avg)
```

```python
# Get data of students who scored at least 80 in both subjects.
data[(data[0:,1] >= 80) & (data[0:,2] >= 80)]
```

```python
# Replace all Science marks < 75 with 0.
data[:,2][data[:,2] < 75] = 0
data
```

# Mini Project: E-Commerce Sales Analytics System

**[Customer_ID, Product_Category, Quantity, Price_Per_Item, Discount(%), Rating]**
Categories:

* 101 → Electronics
* 102 → Clothing
* 103 → Grocery
* 104 → Books

```python
import numpy as np

sales_data = np.array([
    [1001, 101, 2, 5000, 10, 4],
    [1002, 102, 1, 1500, 5, 5],
    [1003, 103, 10, 200, 0, 3],
    [1004, 101, 1, 7000, 15, 4],
    [1005, 104, 5, 300, 0, 5],
    [1006, 102, 3, 1200, 8, 2],
    [1007, 103, 8, 150, 0, 4],
    [1008, 101, 1, 10000, 20, 5],
    [1009, 104, 7, 250, 0, 3],
    [1010, 102, 2, 1800, 5, 4]
])
```

```python
#1. Calculate Final Bill
bill = sales_data[0:,2] * sales_data[0:,3]
bill
```

```python
final_bill = (bill * (1 - (sales_data[0:,4]) / 100))
final_bill = final_bill.astype(int)
final_bill
```

```python
# Find customers whose final bill is greater than ₹5000.
customers = final_bill > 5000
sales_data[customers]
```

```python
Electronics = sales_data[0:,1] == 101
Clothing = sales_data[0:,1] == 102
Grocery = sales_data[0:,1] == 103
Books = sales_data[0:,1] == 104

Electronics = sales_data[Electronics]
Clothing = sales_data[Clothing]
Grocery = sales_data[Grocery]
Books = sales_data[Books]
```

```python
sales_data = np.column_stack((sales_data, final_bill))
sales_data
```

```python
# Find total revenue generated by each category.
print(sum(Electronics[0:,6]), sum(Clothing[0:,6]), sum(Grocery[0:,6]), sum(Books[0:,6]))
```
