# Pandas Series

**A series is a one dimentional labeled array use for holdiing any type of data. The axis labels are also called the index.**

# Creating Series

```python
import numpy as np
import pandas as pd
```

```python
my_list = ['A', 'B', 'C']
my_arr = np.array([1,2,3])
my_dict = {1:10, 2:20, 3:30}
```

```python
print(pd.Series(my_list))
print(pd.Series(my_arr,index=my_list))
print(pd.Series(my_dict))
```

# Missing Data

**Finding Missing Data**

```python
data = {
    "A": [np.nan, 20, 30, 40],
    "B": [10, np.nan, 30, 40],
    "C": [10, 20, 30, 40],
    "D": [10, np.nan, 30, np.nan]
}
df = pd.DataFrame(data)
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

**Removing Missing Data**

```python
df.dropna() # remove all rows whose have NaN value
```

```python
df.dropna(thresh=3) # thresh means number of non zero element present in row
```

**Filling Mising Data**

```python
df
```

df.fillna(0)

```python
values = {'A': 10, 'B' : 20, 'C' : 30, 'D': 40}
df.fillna(value=values)
```

```python
df.fillna(df.mean())
```

# Merging Joining and Concatenation

**Merging**

```python
# Data Frame for Employees
employees = {
    "Emp_ID": [101, 102, 103, 104],
    "Name": ["Dhruv", "Rahul", "Priya", "Neha"],
    "Department_ID": [1, 2, 1, 3]
}
df_employees = pd.DataFrame(employees)
# Data Frame for Departments
departments = {
    "Department_ID": [1, 2, 3],
    "Department_Name": ["AI/ML", "Web Dev", "Cyber Security"]
}
df_departments = pd.DataFrame(departments)
```

```python
df_employees
df_departments
```

```python
pd.merge(df_employees, df_departments, on= 'Department_ID')
```

```python
pd.merge(df_employees, df_departments, on= 'Department_ID', how = 'inner')
```

```python
pd.merge(df_employees, df_departments, on= 'Department_ID', how = 'outer')
```

```python
pd.merge(df_employees, df_departments, on= 'Department_ID', how = 'left') # left means according to df_employees
```

```python
pd.merge(df_employees, df_departments, on= 'Department_ID', how = 'right') # right means according to df_departments
```

**Concatenation**

```python
pd.concat([df_employees,df_departments])
```

**Joining**

```python
df1 = {"Name" : ["Bob", "Alex", "Micheal"]}
df_df1 = pd.DataFrame(df1)
df2 = {"Marks" : [80, 90, 70]}
df_df2 = pd.DataFrame(df2,index=[2,3,4])
```

```python
df_df1.join(df_df2)
```

```python
df_df2.join(df_df1)
```
