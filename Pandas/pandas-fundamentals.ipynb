# Pandas Series

**A series is a one dimentional labeled array use for holdiing any type of data. The axis labels are also called the index.**

# Creating Series

```python
import numpy as np
import pandas as pd
```

```python
my_list = ['A', 'B', 'C']
my_arr = np.array([1,2,3])
my_dict = {1:10, 2:20, 3:30}
```

```python
print(pd.Series(my_list))
print(pd.Series(my_arr,index=my_list))
print(pd.Series(my_dict))
```
