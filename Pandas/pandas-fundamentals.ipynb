# Pandas Series

**A series is a one dimentional labeled array use for holdiing any type of data. The axis labels are also called the index.**

# Creating Series

```python
import numpy as np
import pandas as pd
```

```python
my_list = ['A', 'B', 'C']
my_arr = np.array([1,2,3])
my_dict = {1:10, 2:20, 3:30}
```

```python
print(pd.Series(my_list))
print(pd.Series(my_arr,index=my_list))
print(pd.Series(my_dict))
```

# Pandas DataFrame  

**A data Frame is a collection of multiple Series**

# Creating a Data Frame

```python
# creating a data frame using dictionaries 
student_dict = {
    "ID": [101, 102, 103, 104],
    "Name": ["Dhruv", "Rahul", "Priya", "Neha"],
    "Age": [19, 20, 18, 21],
    "Course": ["AI/ML", "Web Dev", "Data Science", "Cyber Security"]
}
df = pd.DataFrame(student_dict)
df
```

```python
# Craeting a data frame using lists
student_list = [
    [101, "Dhruv", 19, "AI/ML"],
    [102, "Rahul", 20, "Web Dev"],
    [103, "Priya", 18, "Data Science"],
    [104, "Neha", 21, "Cyber Security"]
]

df_list = pd.DataFrame(student_list)
df_list
```

```python
columns = ["ID", "Name", "Age", "Course"]
df_list2 = pd.DataFrame(student_list, columns= columns)
df_list2
```

**Selection and indexing of Columns**

```python
df_list2["Name"]
df_list2[["Name", "Age"]]
```

**Creating a new Column**

```python
df_list2["Salary"] = [20000, 10000, 15000, 12000]
df_list2
```
