# Array Generation Function

```python
arr1 = np.zeros(5)
arr2 = np.zeros((3,3))
print(arr1)
print(arr2)
```

```python
arr1 = np.ones(5)
arr2 = np.ones((5,5))
print(arr1)
print(arr2)
```

```python
arr1 = np.linspace(1,10,5)
print(arr1)
```

# Numpy Indexing and slicing of vectors

```python
# slicing - array[start,stop,steps]
arr = np.arange(21,31)
arr
```

```python
print(arr[4])
print(arr[1:5])
```

```python
print(arr[3:])
print(arr[:5])
print(arr[1::3])
```

```python
print(arr[::]) # print whole array 
print(arr[::-1]) # print whole array in reverse order
```

# Numpy Indexing and slicing of Matrix

```python
arr = np.arange(11,31).reshape(4,5)
arr
```

```python
arr[2,3]
arr[2,0]
```

```python
print(arr[0:2, 1:3])
print(arr[1:, 3:])
```

```python
arr[1:,2]
```

# Boolean indexing

```python
arr = np.arange(1,11)
arr
```

```python
bool_indexing = arr % 2 == 0
bool_indexing
```

```python
arr = arr[bool_indexing]
arr
```
