# Array Operations

**Arithmetic Operations**

```python
arr1 = np.array([1,23,4,3,11,23])
arr2 = np.array([11,32,41,31,21,93])
```

```python
print(arr1 + arr2) # Addition
print(arr2 - arr1) # Subtraction
print(arr1 * arr2) # Multiplication
print(arr1 / arr2) # Divition
print(arr2 // arr1) # Floor divition
print(arr1 ** arr2) # exponent 
```

**Broadcasting**

```python
arr = arr1.reshape(3,2)
```

```python
arr + 10
```

**Deep and shallow copy**

```python
# Shallow copy - whenever we make any changes on an array but its not reflect the original array
# Deep copy - whenever we make any changes on an array but its reflcet the original array means when we modify the array 
#             then original array also modified
```

```python
arr1 = np.array([1,2,3,4,5])
arr2 = np.array([1,2,3,4,5])
sli = arr1[2:]
print(sli)
print(arr1) # arr1 remains same
```

```python
b = arr1.copy() # shallow copy
c = arr2 # deep copy
c[2] = 30
print(b * 10)
print(c + 10)
```

```python
print(arr1)
print(arr2)
```

# Advance Array Manipulation

**Stacking Array**

```python
a1 = np.array([1,2,3,4,5])
a2 = np.array([11,22,33,44,55])
```

```python
np.vstack((a1,a2)) # vertically arrange the values
```

```python
np.hstack((a1,a2)) # horizontally arrange the values
```

```python
np.column_stack((a1,a2)) # column wise arrange the values
```
